```python
import matplotlib.pyplot as plt

t = []
r = []

L=len("LOGGING: Throttle: ")

for line in open("outpout.txt", "r"):
    a,b = line[L:].split(", RPM: ")
    t.append(int(a))
    r.append(int(b))

plt.plot(t,r)
plt.xlabel("Throttle")
plt.ylabel("RPM")
plt.show()
```
